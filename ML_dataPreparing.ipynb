```python
import numpy as np
import pandas as pd
import sklearn
import pandas_datareader.data as web

    

df_raw = web.DataReader(name='^SPX', data_source='stooq',start='2019-09-10', end='2022-09-08')
df_raw.head()

```

```python
df_copy=df_raw.copy()
df=df_copy[:]
df.info()
df['day']=df.index.day
df['month']=df.index.month
df['year']=df.index.year
df['value_class_close']=pd.cut(df.Close,bins=5)
df['value_class_open']=pd.cut(df.Open,(2000,2500,3000,3500,4500),labels=['extreme_low','low','medium','high'])
df
```

```python
df=pd.DataFrame(data={"height":[168.,170.,185.,187.,176.,145.,196]})
df["category_intervals"]=pd.cut(x=df.height,bins=3)
df["category_category"]=pd.cut(x=df.height,bins=(140,170,175,200), labels=['low','medium','high'])
pd.get_dummies(df,drop_first=True,prefix='height')
```

```python
df=pd.DataFrame(data={"lang":[['PL','ENG'],['GER','ENG','PL','FRA'],['RUS']]})
df["lang_number"]=df["lang"].apply(len)
df["IS_POLISH"]=df["lang"].apply(lambda x:1 if 'PL' in x else 0)
df
```

```python
df=pd.DataFrame(data={"website":['wp.pl','one.pl','google.com']})
new=df.website.str.split(".",expand=True)
df["extention"]=new[1]
df
```
