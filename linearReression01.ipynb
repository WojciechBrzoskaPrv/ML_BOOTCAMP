```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
np.set_printoptions(precision=6, suppress=True, edgeitems=30, linewidth=120, 
                    formatter=dict(float=lambda x: f'{x:.2f}'))
sns.set(font_scale=1.3)
sklearn.__version__
```

```python
from sklearn.datasets import make_regression

data, target = make_regression(n_samples=100, n_features=1, n_targets=1, noise=30.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```python
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
our_estimation=LinearRegression()
met_attr=[item for item in dir(our_estimation) if not item[:1]=='_']
print(met_attr,'\n')
our_estimation.fit(data,target)
our_estimation.get_params()
R2=our_estimation.score(data,target)
print('{:.2f}x+{:.2f} with R2={:.2f}%'.format(our_estimation.coef_[0],our_estimation.intercept_,R2*100.00))
regEqStr='{:.2f}x+{:.2f} with R2={:.2f}%'.format(our_estimation.coef_[0],our_estimation.intercept_,R2*100.00)
print(regEqStr)
modelValues=our_estimation.predict(data)
```

```python
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa:'+regEqStr,{'fontsize':12})
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.plot(data,modelValues,color='red')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.show()
```
